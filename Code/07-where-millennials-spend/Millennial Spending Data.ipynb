```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mil = pd.read_csv('millennial-spending-update.csv')
mil
```

```python
mil = mil[['Unnamed: 0', 'All households', 'Greatest Gen', 'Silent Gen', 'Baby Boomers', 'Generation X', 'Millenials']]
```

```python
mil
```

```python
mil.to_csv("ordered-millennials.csv", index=False)
```

```python
mil.rename(columns={'Unnamed: 0': 'Category'}, inplace=True)
```

```python
#Greatest: Born 1927 or earlier
#Silent: 1928–1945
#Baby Boomers: 1946–1964
#Generation X: 1965–1980
#Millenials: 1981–present
mil
```

```python
food = mil.loc[mil['Category'].isin(['Food at home', "Food away from home"])]
food
```

```python
ax = food.plot(kind='bar', legend=True, figsize=(10, 6))
plt.legend(loc="upper right")
plt.style.use('ggplot')

ax.set_title("Total Food Expenditures By Generation", pad=40, fontsize=20)
plt.gcf().text(0.5, .93, "", ha='center', fontsize=10)

ax.set_xlabel('Type of Food Expense')
ax.set_ylabel('Percentage of Total Household Budget')
ax.set_ymargin(10)

plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42
})

plt.savefig("food-expense.pdf")
plt.show()
```

```python
housing = mil.loc[mil['Category'].isin(['Owned dwellings', "Rented dwellings"])]
housing
```

```python
ax = housing.plot(kind='bar', legend=True, figsize=(10, 6))
plt.legend(loc="upper right")
plt.style.use('ggplot')

ax.set_title("Total Housing Expenditures By Generation", pad=40, fontsize=20)
plt.gcf().text(0.5, .93, "", ha='center', fontsize=10)

ax.set_xlabel('Type of Housing Expense')
ax.set_ylabel('Percentage of Total Household Budget')

plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42
})

plt.savefig("housing-expense.pdf")
plt.show()
```

```python
cars = mil.loc[mil['Category'].isin(['New cars and trucks', "Used cars and trucks"])]
cars
```

```python
ax = cars.plot(kind='bar', legend=True, figsize=(10, 6))
plt.legend(loc="upper right")
plt.style.use('ggplot')

ax.set_title("Buying a Car By Generation in 2016", pad=40, fontsize=20)
plt.gcf().text(0.5, .93, "", ha='center', fontsize=10)

ax.set_xlabel('Type of Car Purchased')
ax.set_ylabel('Percentage of Total Household Budget')

plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42
})

plt.savefig("car-expense.pdf")
plt.show()
```

```python
ax = cars.plot(kind='line', legend=True, figsize=(10, 6))
plt.legend(loc="upper right")
plt.style.use('ggplot')

ax.set_title("Buying a Car By Generation in 2016", fontsize=10)
plt.gcf().text(0.5, .93, "", ha='center', fontsize=10)

ax.set_xlabel('Type of Car Purchased')
ax.set_ylabel('Percentage of Total Household Budget')

plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42
})

plt.savefig("car-lines.pdf")
plt.show()
```
